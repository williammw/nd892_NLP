```python
r = requests.get("https://hk.finance.yahoo.com/")
print(r.text)
```

```python
import re

pattern = re.compile(r'<.*?>')
print(pattern.sub('',r.text))
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text,"html5lib")
print(soup.get_text())
```

```python
#extract title
header_link = soup.find_all("li", class_="js-stream-content")
#header_link[0].select_one("a").get_text().strip()
header_link
```

```python
# Extract description
header_link = soup.find_all("div", class_="o-teaser--article")
for link in header_link:
    title = link.select_one("a").get_text().strip()
    description = link.select_one("js-teaser-standfirst-link").get_text().strip()
    print("***", title, "***")
    print(description)
```
